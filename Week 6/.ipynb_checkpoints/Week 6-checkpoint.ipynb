```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_groq import ChatGroq
from langchain.chains import RetrievalQA
from langchain.docstore.document import Document
```

```python
with open('Youghurt.txt', 'r') as f:
    text = f.read()

paragraphs = text.split("\n\n")  # split by double newlines
docs = [Document(page_content=para.strip()) for para in paragraphs if para.strip()]
```

```python
# print(docs)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)

split_docs = text_splitter.split_documents(docs)
```

```python
# print(split_docs)
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
vectorstore = Chroma.from_documents(
    split_docs,
    embedding_model
)

retriever = vectorstore.as_retriever(search_kwargs={"k": 3})
```

```python
import os
os.environ["GROQ_API_KEY"] = "Your Groq API Key"
```

```python
llm = ChatGroq(
    model_name="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    return_source_documents=True
)
```

```python
query = "How long does youghurt preparation take?"
response = qa_chain(query)

print("Answer:")
print(response["result"])

print("\nSources:")
for doc in response["source_documents"]:
    print("-", doc.page_content)
```

```python
test_cases = [
    {
        "id": "TC001",
        "query": "What temperature should milk be heated to?",
        "expected": "Milk should be heated to around 82 to 85 degrees Celsius."
    },
    {
        "id": "TC002",
        "query": "How long should yogurt incubate?",
        "expected": "Yogurt should incubate for 6 to 12 hours."
    },
    {
        "id": "TC003",
        "query": "How do you make yogurt thicker?",
        "expected": "You can add powdered milk or strain it to make Greek yogurt."
    },
    {
        "id": "TC004",
        "query": "What happens if milk is too hot?",
        "expected": "It can kill the bacteria in the starter culture."
    }
]
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

def similarity_score(expected, actual):
    emb1 = model.encode(expected, convert_to_tensor=True)
    emb2 = model.encode(actual, convert_to_tensor=True)
    return util.cos_sim(emb1, emb2).item()
```

```python
response = qa_chain(query)
```

```python
def get_chatbot_answer(query):
    response = qa_chain(query)
    
    # Adjust depending on your output format
    if isinstance(response, dict):
        return response.get("result", "")
    
    return str(response)
```

```python
results = []

for test in test_cases:
    query = test["query"]
    expected = test["expected"]
    
    actual = get_chatbot_answer(query)
    
    score = similarity_score(expected, actual)
    
    if score >= 0.85:
        result = "Pass"
    elif score >= 0.7:
        result = "Partial"
    else:
        result = "Fail"
    
    results.append({
        "id": test["id"],
        "query": query,
        "expected": expected,
        "actual": actual,
        "score": score,
        "result": result
    })
```

```python
import pandas as pd

df = pd.DataFrame(results)
df
```

```python
total = len(df)
passed = len(df[df["result"] == "Pass"])

accuracy = passed / total * 100

print(f"Accuracy: {accuracy:.2f}%")
```

```python
def llm_judge(query, expected, actual):
    prompt = f"""
    Question: {query}
    Expected Answer: {expected}
    Chatbot Answer: {actual}
    
    Does the chatbot answer correctly answer the question?
    Reply only with YES or NO.
    """
    
    return llm.invoke(prompt)
```
